```python
import pandas as pd
df=pd.read_csv('/content/AB_NYC_2019.csv')
df
```

```python
df.info()
```

```python
#Tìm giá trị null
df.isnull().sum()
```

```python
#Xóa các giá trị null, bị trùng lập
df=df.dropna()
df = df.drop_duplicates( keep=False)
print(df)
```

```python
#Xóa các cột không cần thiết
df = df.drop('calculated_host_listings_count', axis=1)
df.drop(['id', 'last_review'], axis = 1, inplace = True)
print(df)
```

```python
df.info()
```

```python
#Xuất dữ liệu đã chuẩn hóa
df.to_csv('/content/AB_NYC_2019.csv',
          columns = ['name' ,	'host_id' ,	'host_name' ,	'neighbourhood_group' ,	'neighbourhood' ,	'latitude' ,	'longitude' ,	'room_type' ,	'price' ,	'minimum_nights' ,	'number_of_reviews'
          , 'reviews_per_month' ,	'availability_365' ]
          , sep='\t', index = False)
```
